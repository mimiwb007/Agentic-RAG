```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_ollama.llms import OllamaLLM
from IPython.display import Markdown

template = """Question: {question}

Answer: Let's think step by step."""

prompt = ChatPromptTemplate.from_template(template)

model = OllamaLLM(model="granite3-moe:3b")

chain = prompt | model

# Prepare input for invocation
input_data = {
    "question": 'Extract all numerical values and their corresponding units from the text: "The marathon was 42 kilometers long, and over 30,000 people participated."'}

# Invoke the chain with input data and display the response in Markdown format
response = chain.invoke(input_data)
display(Markdown(response))
```
